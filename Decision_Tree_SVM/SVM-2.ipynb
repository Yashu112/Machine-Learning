Q1. What is the relationship between polynomial functions and kernel functions in machine learning
algorithms?

Ans. Polynomial functions and kernel functions are both mathematical tools used in machine learning, particularly in the context of support vector machines (SVMs) and kernel methods. They are related in the sense that kernel functions can be used to implicitly map data into higher-dimensional spaces, making it possible to learn complex, non-linear relationships between features. Polynomial functions are one type of kernel function.

Polynomial Kernel:
   - A polynomial kernel is a specific type of kernel function used in SVMs and other kernel-based algorithms.
   - It is defined as $K(x, y) = (x_1 \cdot x_2  + c)^d$, where x_1 and x_2 are data points, c is a constant, and d is the degree of the polynomial.
   - The polynomial kernel computes the inner product of the data points x and y in an implicit high-dimensional space without explicitly mapping the data to that space.

   - Polynomial functions are used as kernel functions in SVMs and kernel methods to capture non-linear relationships in the data.
   - By adjusting the degree parameter (d) and the constant (c) in the polynomial kernel, you can control the degree of non-linearity in the decision boundary learned by the SVM.


Q2. How can we implement an SVM with a polynomial kernel in Python using Scikit-learn?

Ans. Solution using python is as follows:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification

X,y=make_classification(n_samples=1000, n_features=2, n_classes=2,
                       n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.25, random_state=42)

from sklearn.svm import SVC
svc=SVC(kernel='poly')

svc.fit(X_train, y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(f'confusion Matrix: \n{confusion_matrix(y_pred=y_pred,y_true=y_test)}\n')
print(f'Accuracy score : {accuracy_score(y_pred,y_test)}\n')
print(classification_report(y_pred,y_test))
```

Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?

Ans. Increasing the value of epsilon $(\epsilon)$ in Support Vector Regression (SVR) typically leads to an increase in the number of support vectors. Support vectors are data points that lie on or inside the margin (i.e., within a distance of ε) and have a significant influence on determining the SVR's regression function. When $(\epsilon)$ is larger, the margin becomes wider, and more data points can fall within this wider margin, thus becoming support vectors. Consequently, as $(\epsilon)$ increases, the number of support vectors tends to increase as well. This can result in a simpler and more flexible regression model, but it may also lead to reduced accuracy in capturing fine-grained details in the data, as the model becomes less sensitive to small variations.

Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter
affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works
and provide examples of when you might want to increase or decrease its value?

Ans. The performance of Support Vector Regression (SVR) is significantly affected by various parameters, including the choice of kernel function, the $C$ parameter, the epsilon $\epsilon$ parameter, and the gamma $\gamma$ parameter. 

1. **Kernel Function:**
   - The kernel function determines how SVR models the non-linear relationships in the data.
   - Common kernel functions include linear, polynomial, and radial basis function (RBF) kernels.
   - Choice of Kernel:
     - Linear kernel (no transformation) is suitable when you believe the relationship between input features and output is approximately linear.
     - Polynomial kernel with a higher degree is useful when you expect higher-order polynomial relationships in the data.
     - RBF kernel is a good choice for capturing complex, non-linear relationships, but it can be sensitive to the gamma parameter.

2. **C Parameter:**
   - The C parameter controls the trade-off between maximizing the margin and minimizing the training error.
   - Small C values result in a wider margin but may allow more training errors.
   - Large C values make the margin narrower but enforce a strict fit to the training data.
   - Adjusting C:
     - Increase C when you want the SVR to closely fit the training data, especially when you believe the data has low noise and outliers.
     - Decrease C when you want a smoother, more generalized model that tolerates some training errors.

3. **Epsilon ($\epsilon$) Parameter:**
   - Epsilon defines the margin of tolerance around the predicted values within which no penalty is associated with errors.
   - It sets the size of the tube around the regression line.
   - Adjusting $\epsilon$:
     - Increase $\epsilon$ when you want to allow larger deviations between predicted and actual values in the training data.
     - Decrease $\epsilon$ when you want to enforce a stricter fit to the training data, which can make the model more sensitive to outliers.

4. **Gamma ($\gamma$) Parameter:**
   - The gamma parameter influences the shape and flexibility of the decision boundary in the case of the RBF kernel.
   - It controls the extent of influence of a single training example, with smaller values making it more influential.
   - Adjusting $\gamma$:
     - Increase $\gamma$ when you want to make the SVR more sensitive to local variations in the data, potentially leading to overfitting.
     - Decrease $\gamma$ when you want the SVR to have a smoother and more global fit to the data.

It's important to note that the optimal values for these parameters depend on the specific dataset and problem you are working on. Hyperparameter tuning techniques like cross-validation can help you find the best combination of kernel function, C, ε, and γ for your SVR model. Additionally, regularization techniques such as grid search or randomized search can assist in selecting appropriate parameter values. The choice of parameter values should aim to strike a balance between model complexity, generalization, and robustness to noise and outliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('penguins')
```

```python
df.info()
```

Here we will use the Penguins dataset and make the sex column as our target an try to predict whether the penguin is male or female

```python
df.isnull().sum()
```

```python
df.drop(index=np.where(df['sex'].isnull())[0], axis=1, inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

No null values now in the dataset

```python
df.nunique()
```

We need to encode species, island columns

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

encoded=encoder.fit_transform(df[['species', 'island']]).toarray()
encoded_df=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
df=pd.concat([df,encoded_df], axis=1)
df
```

```python
df.drop(columns=['species', 'island'], inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split

columns=df.columns
X=df[columns[columns!='sex']]
y=df['sex']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=12)

X_train.shape, X_test.shape
```

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(f'confusion Matrix: \n{confusion_matrix(y_pred=y_pred,y_true=y_test)}\n')
print(f'Accuracy score : {accuracy_score(y_pred,y_test)}\n')
print(classification_report(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={'C':[0.1,1,0.5,15],
            'kernel':['poly','rbf'],
}

clf=GridSearchCV(SVC(), param_grid=param_grid, cv=5)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
new_svc = SVC(C=15, kernel='rbf')

new_svc.fit(X_train,y_train)
y_pred=new_svc.predict(X_test)
```

```python
print(f'confusion Matrix: \n{confusion_matrix(y_pred=y_pred,y_true=y_test)}\n')
print(f'Accuracy score : {accuracy_score(y_pred,y_test)}\n')
print(classification_report(y_pred,y_test))
```

Now, saving the tuned model in pickle file:
