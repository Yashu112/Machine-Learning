Q1. What is the mathematical formula for a linear SVM?

Ans. The mathematical formula for a linear SVM is:

$$ y = w^Tx + b $$

where, 
- y is target
- x is the feature matrix
- w is the co-efficient matrix
- b is linear coefficient estimated from data

Q2. What is the objective function of a linear SVM?

Ans. The objective function for a linear SVM can be formulated as follows:

$$ \text{Minimize } \frac{1}{2} ||w||^2 $$

subject to the constraint:

$$y_i*(w^T \cdot x_i + b) \geq 1$$

for all training samples $(x_i, y_i)$, where:
- $y_i$ is the class label of the $i$-th sample (either +1 or -1 for binary classification).
- $w^T \cdot x_i + b$ is the decision function that assigns a score to each data point.

The constraint enforces that all data points should be correctly classified and lie outside a certain margin (a distance of at least 1 from the decision boundary).



Q3. What is the kernel trick in SVM?

Ans. The kernel trick is a fundamental concept in Support Vector Machines (SVMs) and is used to handle non-linearly separable data by implicitly mapping it into a higher-dimensional feature space without explicitly computing the transformation. It allows SVMs to perform well on complex classification problems that cannot be effectively solved using linear separation.

The primary idea behind the kernel trick is to find a function, known as a kernel function, that computes the dot product between the transformed feature vectors in the higher-dimensional space without actually computing the transformation explicitly. This allows SVMs to effectively work in a high-dimensional space without the computational burden of explicitly transforming the data.

Mathematically, in a linear SVM, the decision boundary is defined as:

$$ y = w^Tx + b $$

Where:
- $f(x)$ is the decision function.
- $w$ is the weight vector.
- $x$ is the input feature vector.
- $b$ is the bias term.

The kernel trick introduces the kernel function $K(x, x')$ that calculates the dot product between the transformed feature vectors in a higher-dimensional space:

Some commonly used kernel functions include:

1. **Linear Kernel**: $ K(x, x') = x \cdot x'$
2. **Polynomial Kernel**: $K(x, x') = (x \cdot x' + c)^d $
3. **Radial Basis Function (RBF) Kernel (Gaussian Kernel)**: $ K(x, x') = \exp\left(-\frac{||x - x'||^2}{2\sigma^2}\right)$

Q4. What is the role of support vectors in SVM Explain with example

Ans. Support vectors play a crucial role in Support Vector Machines (SVMs) as they are the data points that are closest to the decision boundary (hyperplane) and have a direct influence on defining the position and orientation of the hyperplane. These support vectors are the most informative data points in the dataset, and they determine the maximum margin between the classes. 

**Example**:

Consider a binary classification problem with two classes, represented by red and blue dots in a 2D feature space. The goal is to find a hyperplane that separates these two classes as effectively as possible.



In this example, the solid black line represents the decision boundary (hyperplane) found by the SVM. The two dashed lines parallel to the decision boundary represent the margins. The support vectors are the data points that are closest to these margins, specifically the points on the margins' edges.

The role of support vectors can be explained as follows:

1. **Determining the Margin**: The distance between the decision boundary and the margin is called the margin width. The SVM's objective is to maximize this margin while correctly classifying all data points. The support vectors are the data points that are closest to the margin, and they define the margin's width. Maximizing this margin ensures that the decision boundary is optimally positioned between the classes.

2. **Supporting the Decision Boundary**: The support vectors are the most influential data points when it comes to defining the position and orientation of the decision boundary. In essence, the decision boundary is determined by these support vectors, and any changes in their positions could potentially alter the decision boundary.

3. **Robustness**: Support vectors are crucial for the SVM's robustness. They are the data points that are most difficult to classify or are closest to the decision boundary. By focusing on these critical points, the SVM aims to make the classification as robust as possible to variations in the data.


Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

Ans. 
**1. Hyperplane**: The hyperplane is the decision boundary that separates two classes in an SVM. In a binary classification problem, it's a flat, high-dimensional surface. The goal is to find a hyperplane that maximizes the margin between the classes.

**2. Marginal Plane**: The marginal plane is the region defined by the margin around the hyperplane. It includes all the points within a certain distance from the hyperplane.

**4. Hard Margin**: A hard margin SVM requires that all data points are correctly classified, with no tolerance for misclassification. It's suitable when the data is linearly separable and noise-free.


**Example Hyperplane, Marginal plane, Hard margin:**

Consider a simple 2D classification problem with two classes, red and blue circles. Here's a graph illustrating the hyperplane that separates them. In this case, the black line is the hyperplane. It effectively separates the red and blue circles in 2D space.

The region between the dashed lines is the marginal plane. It includes all the points within a certain distance from the hyperplane.


In this example, all points are correctly classified, and there is no misclassification allowed. The margin is maximized without any data points falling inside it. So, it is example of hard margin.





**3. Soft Margin**:

In real-world scenarios, data is often not perfectly separable, and some misclassification might be tolerated. In such cases, soft margin SVMs allow for some data points to be on the wrong side of the margin or even on the wrong side of the hyperplane, but with a penalty.

Q6. SVM Implementation through Iris dataset.

- Load the iris dataset from the scikit-learn library and split it into a training set and a testing set
- Train a linear SVM classifier on the training set and predict the labels for the testing set
- Compute the accuracy of the model on the testing set
- Plot the decision boundaries of the trained model using two of the features
- Try different values of the regularisation parameter C and see how it affects the performance of
    the model.
    


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris

dataset=load_iris()

X=pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
y=pd.DataFrame(dataset.target)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=25)

X_train.head()
```

```python
from sklearn.svm import SVC

svc = SVC()

svc.fit(X_train, y_train)

y_pred=svc.predict(X_test)
print('The predicted values are:')
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(f'confusion Matrix: \n{confusion_matrix(y_pred=y_pred,y_true=y_test)}\n')
print(f'Accuracy score : {accuracy_score(y_pred,y_test)}\n')
print(classification_report(y_pred,y_test))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

svc.fit(X_train.iloc[:,:2],y_train)

disp = DecisionBoundaryDisplay.from_estimator(
        svc,
        X.iloc[:,:2],
        response_method="predict",
        alpha=0.8,  
    )
```

```python
from sklearn.model_selection import GridSearchCV

parameter={'C': list(np.linspace(0,1, num=25))}

clf=GridSearchCV(SVC(), param_grid=parameter, cv=5, verbose=2.0)

clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
C_values = clf.cv_results_['param_C'].data
mean_scores = clf.cv_results_['mean_test_score']
```

```python
sns.lineplot(x=C_values, y=mean_scores)
plt.xlabel('C Values')
plt.ylabel('Corresponding Scores')
plt.title('Plot of Different C values and the resulting scores')
plt.show()
```

Thus, we see that as C value increases, the sccuracy increases and reaches a maximum before 1 and again drops a little
