Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.

Ans. Simple Linear Regression and Multiple Linear Regression are two types of regression analysis used in statistics to model the relationship between a dependent variable and one or more independent variables. They have distinct differences:

1. Simple Linear Regression:
   - Simple Linear Regression (SLR) is used when there is a linear relationship between a dependent variable (y) and a single independent variable (X).
   - It aims to find a linear equation that best fits the data and predicts Y based on changes in X.
   - The equation for the best fit line predicted using simple linear regression is represented as:
   
     $$h_{\theta}(x) = \theta_0 + \theta_1{x_1}$$
     
     where $h_{\theta}(x)$ is predicted value of the dependent variable, $x_1$ is the independent variable, $\theta_0$ is the intercept, and $\theta_1$ is the slope of the line.
   - Example: Suppose we want to predict a person's weight ($h_{\theta}(x)$) based on their height ($x_1$). In this case, height is the only independent variable.

2. Multiple Linear Regression:
   - Multiple Linear Regression (MLR) is used when there is a linear relationship between a dependent variable (y) and two or more independent variables $(x_1, x_2, x_3, ...)$.
   - It extends the concept of SLR to incorporate multiple independent variables.
   - The equation for MLR is represented as:
   
     $$h_{\theta}(x) = \theta_0 + \theta_1{x_1} +  \theta_2{x_2} + .... +  \theta_n{x_n}$$
     
    where $h_{\theta}(x)$ is predicted value of the dependent variable, $x_1, x_2, x_3, ...$ are the independent variable, $\theta_0$ is the intercept, and $\theta_1$, $\theta_2$, etc are the coefficients for each independent variable.
    
   - Example: Suppose we want to predict a car's gas mileage ($h_{\theta}(x)$) based on factors like engine size ($x_1$), weight ($x_2$), and horsepower ($x_3$). In this case, there are multiple independent variables.


Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in
a given dataset?

Ans. Linear regression makes several key assumptions about the data and the model. These assumptions are important because violations of these assumptions can lead to unreliable and biased results. Here are the common assumptions of linear regression:

- Linearity: The relationship between the independent variables and the dependent variable should be linear. This means that changes in the independent variables should lead to proportional changes in the dependent variable. 

    We can check this assumption by creating scatterplots of the independent variables against the dependent variable and looking for a roughly linear pattern.

- Independence of Errors: The errors (residuals), which are the differences between the observed and predicted values of the dependent variable, should be independent of each other. Autocorrelation in residuals can indicate a violation of this assumption. 

    We can use residual plots or autocorrelation plots to check for this.

- Constant Variance: The variance of the errors should be constant across all levels of the independent variables. In other words, the spread of residuals should remain relatively constant as we move along the range of the independent variables. 

    We can check for this by plotting residuals against predicted values or independent variables and looking for patterns or a cone-shaped spread.

- Normality of Errors: The errors should be normally distributed. This assumption is important for making inferences and constructing confidence intervals. 

    We can check this assumption using histograms, Q-Q plots, or statistical tests like the Shapiro-Wilk test.

- No or Little Multicollinearity: In multiple linear regression, the independent variables should not be highly correlated with each other (multicollinearity). High multicollinearity can make it difficult to separate the individual effects of the independent variables. 

    We can assess multicollinearity using correlation matrices and heatmap of correlation matrix.

Q3. How do you interpret the slope and intercept in a linear regression model? Provide an example using
a real-world scenario.

Ans. In any linear regression model with equation : $h_{\theta}(x) = \theta_0 + \theta_1{x_1} +  \theta_2{x_2} + .... +  \theta_n{x_n}$

- $\theta_0$ denotes the **intercept**. It is the value of the dependent variable when the value of all independent variables is zero

- $\theta_1$, $\theta_2$, etc denote the **slope** of variables. It is equal to differential of dependent variable with respect to independent variable. The slope signifies the amount of change in dependent variable when there is a unit change in the independent variable.

For example, consider some random Experience-Salary data which is fitted and plotted in the form of a straight line. This is an example of simple linear regression. Here Salary is dependent variable and experience is independent variable.

Now, in this example, the intercept denotes the salary even when experience is 0. And slope here denotes the average increase in salary per year.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

experience = list(range(0,12,2))
salary = [25000,30000,35000,40000,46000,60000]

ax=plt.axis([0, 12, 20000, 65000])
sns.regplot(x=experience, y=salary, truncate=False)
plt.show()
```

Q4. Explain the concept of gradient descent. How is it used in machine learning?

Ans. Gradient descent is an iterative optimization algorithm used in machine learning and mathematical optimization to find the minimum of a function. It is particularly important in machine learning for training models, such as linear regression, logistic regression, neural networks, and many others.

**Objective Function:** In machine learning, we often have a cost or loss function that quantifies how well a model's predictions match the actual data. The goal is to minimize this function to make the model's predictions as accurate as possible.

**Gradient:** The gradient of a function is a vector that points in the direction of the steepest increase in the function's value. In the context of optimization, we want to move in the opposite direction of the gradient to find the minimum of the function.

**Algorithm:** Gradient descent starts with an initial guess for the model's parameters (e.g., weights in a linear regression model) and iteratively updates these parameters to minimize the cost function. The algorithm proceeds as follows:

- Compute the gradient of the cost function with respect to the model parameters. This gradient tells us the direction in which the parameters should be adjusted to reduce the cost.
- Update the model parameters by taking a small step (controlled by a hyperparameter called the learning rate) in the opposite direction of the gradient.
- Repeat the above two steps until a stopping criterion is met, such as a maximum number of iterations or when the change in the cost function becomes sufficiently small.
- The algorithm converges to the minimum of the cost function, where the gradient is (approximately) zero.

The convergence algorithm can be written as:

Repeat until convergence:

>j=0,1

>$\theta_j = \theta_j - \alpha . \frac{\partial}{\partial{\theta_j}}J(\theta_0,\theta_1)$

Where 
- $J(\theta_0,\theta_1)$ is our cost/loss function, for example it can be the Mean Squared Error. 
- $\alpha$ is the learning rate



Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?

Ans. Multiple Linear Regression (MLR) is used when there is a linear relationship between a dependent variable (y) and two or more independent variables $(x_1, x_2, x_3, ...)$. It extends the concept of Simple Linear Regression to incorporate multiple independent variables.

The equation for MLR is represented as:
$$h_{\theta}(x) = \theta_0 + \theta_1{x_1} +  \theta_2{x_2} + .... +  \theta_n{x_n}$$
     
where $h_{\theta}(x)$ is predicted value of the dependent variable, $x_1, x_2, x_3, ...$ are the independent variable, $\theta_0$ is the intercept, and $\theta_1$, $\theta_2$, etc are the coefficients for each independent variable.

Example: Suppose we want to predict a car's gas mileage ($h_{\theta}(x)$) based on factors like engine size ($x_1$), weight ($x_2$), and horsepower ($x_3$). In this case, there are multiple independent variables.

The main difference between Simple Linear Regression and Multiple Linear Regression is the number of independent variables. In Simple Linear Regression we have only 1 independent feature $x_1$ but in Multiple Linear Regression we have more than one independent feature as $x_1$, $x_2$, etc.

Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?

Ans. 

**1. Explanation of Multicollinearity:**
   - In multiple linear regression, the goal is to understand how different independent variables jointly influence the dependent variable.
   - When independent variables are highly correlated, it becomes difficult to separate their individual effects because they tend to move together in response to changes in the dependent variable.
   - This can result in unstable coefficient estimates, large standard errors, and inflated p-values, making it challenging to draw meaningful conclusions from the regression analysis.

**2. Detecting Multicollinearity:**
   There are several methods to detect multicollinearity:

   - **Correlation Matrix:** Calculate the correlation coefficients between pairs of independent variables. High absolute correlation coefficients (e.g., > 0.7 or 0.8) indicate potential multicollinearity.
   
   - **Variance Inflation Factor (VIF):** Calculate the VIF for each independent variable. The VIF quantifies how much the variance of the estimated regression coefficient is increased due to multicollinearity. VIF values greater than 1 (typically, a threshold of 5 or 10 is used) are indicative of multicollinearity.

**3. Addressing Multicollinearity:**
   Once multicollinearity is detected, there are several strategies to address it:

   - **Remove Redundant Variables:** If two or more variables are highly correlated and convey similar information, consider removing one of them from the model to reduce multicollinearity.
   
   - **Combine Variables:** Sometimes, you can create a new variable that combines the information from highly correlated variables. For example, if you have height and weight as independent variables that are highly correlated, you can create a body mass index (BMI) variable to replace them.
   
   - **Regularization Techniques:** Ridge regression and Lasso regression are regularization techniques that can help mitigate multicollinearity by adding a penalty term to the regression equation. These methods can force the coefficients of correlated variables to shrink, reducing their impact on the model.

   - **Principal Component Analysis (PCA):** PCA can be used to transform the original set of correlated independent variables into a set of orthogonal (uncorrelated) variables. We can then use these transformed variables in your regression analysis.

   - **Collect More Data:** In some cases, collecting more data may help reduce the impact of multicollinearity. However, this may not always be feasible or effective.


Q7. Describe the polynomial regression model. How is it different from linear regression?

Ans. In Linear Regression we don't use higher power of dependent  variable. We only use power upto 1. The General Equation of Linear Regression is:$$h_{\theta}(x) = \theta_0 + \theta_1{x_1} +  \theta_2{x_2} + .... +  \theta_n{x_n}$$

But in Polynomial Regression, we use higher power of x also in our equation. The linear regression can fit only linear relationships which are represented in the form of straight lines. But it cannot fit non-linear relationships. So, we require Polynomial Regression. So that we can fit polynomials, which are in the form of curves.

The formula for simple polynomial regression for fitting a polynomial of degree 2 is :

$$h_{\theta}(x) = \theta_0 + \theta_1{x_1} +  \theta_2{x_1^2} $$

In general for simple polynomial regression for n-degree polynomial:

$$h_{\theta}(x) = \theta_0 + \theta_1{x_1} +  \theta_2{x_1^2} + \theta_2{x_1^3} + .... +  \theta_n{x_1^n}$$

For Multiple Polynomial Regression, the formula in case of two variables, for degree-2 is:

$$h_{\theta}(x) = \theta_0 + \theta_1{x_1} + \theta_2{x_2} +  \theta_3{x_1^2} +  \theta_4{x_2^2} $$

In general for multiple polynomial regression for n-degree polynomial:

$$h_{\theta}(x) = \theta_0 + \theta_1{x_1} + \theta_2{x_2} + ... + \theta_n{x_n} +  \theta_{n+1}{x_1^2} + \theta_{n+2}{x_1^3} + ... +  \theta_{2n}{x_n^n}$$

Q8. What are the advantages and disadvantages of polynomial regression compared to linear regression? In what situations would you prefer to use polynomial regression?

Ans. The advantages and disadvantages of polynomial regression compared to linear regression are as follows:

**Advantages:**
- We can fit more complex non linear relationship between independent and dependent features.
- Flexibility: By choosing an appropriate degree for the polynomial (e.g., quadratic, cubic, etc.), we can model a wide range of functions, from simple curves to more complex shapes.
- Higher Fit Accuracy: In cases where the underlying relationship is genuinely non-linear, polynomial regression can provide a better fit to the data compared to linear regression, leading to more accurate predictions.

**Disadvanteages:**
- Overfitting: Polynomial regression models with high degrees can be prone to overfitting, where the model fits the training data very closely but may not generalize well to unseen data. It can capture noise in the data instead of the underlying pattern.
- Increased Complexity: Higher-degree polynomial regression models have more parameters, making them computationally expensive and harder to interpret. They can lead to multicollinearity and instability in coefficient estimates.
- Extrapolation Issues: Extrapolating beyond the range of the observed data points can be problematic with polynomial regression, as the model may produce unrealistic predictions outside the observed range

Polynomial regression is preferred in the following situations:

- Non-Linear Relationships: When we believe that the relationship between the independent and dependent variables is non-linear, polynomial regression can be a suitable choice.
- Complex Patterns: When the data exhibits complex, curved patterns that cannot be adequately captured by a linear model, polynomial regression may provide a better fit.
- Exploratory Analysis: Polynomial regression can be useful in exploratory data analysis when we are not sure about the true nature of the relationship and want to test different models.
- Controlled Overfitting: When using polynomial regression, it's essential to be cautious about overfitting. Regularization techniques (e.g., ridge or lasso regression) can be employed to control overfitting by penalizing the complexity of the model.
